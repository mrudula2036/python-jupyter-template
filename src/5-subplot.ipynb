### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

try:
    # Data for the plots
    x = np.linspace(1, 10, 10)
    y = x ** 2

    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]

    x_scatter = [1, 2, 3, 4, 5]
    y_scatter = [2, 4, 6, 8, 10]

    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]

    # Check data integrity properly
    if len(x) == 0 or len(y) == 0:
        raise ValueError("X and Y values for the line plot cannot be empty")
    if len(categories) == 0 or len(values) == 0:
        raise ValueError("Categories and values for the bar chart cannot be empty")
    if len(x_scatter) == 0 or len(y_scatter) == 0:
        raise ValueError("Scatter plot X and Y values cannot be empty")
    if len(labels) == 0 or len(sizes) == 0:
        raise ValueError("Labels and sizes for the pie chart cannot be empty")
    if len(labels) != len(sizes):
        raise ValueError("The number of labels must match the number of sizes")

    # Create a 2x2 subplot layout
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))

    # Line chart
    axs[0, 0].plot(x, y, color='blue', linewidth=2)
    axs[0, 0].set_title("Line Chart")

    # Bar chart
    axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
    axs[0, 1].set_title("Bar Chart")

    # Scatter plot
    axs[1, 0].scatter(x_scatter, y_scatter, color='blue', marker='o')
    axs[1, 0].set_title("Scatter Plot")

    # Pie chart
    axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, 
                  colors=['blue', 'red', 'green', 'orange'], wedgeprops={'edgecolor': 'black'})
    axs[1, 1].set_title("Pie Chart")

    # Adjust layout
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"Error: {e}")
```
