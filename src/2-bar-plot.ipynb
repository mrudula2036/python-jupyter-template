### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt

# Data for the bar chart
try:
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    
    if len(categories) == 0 or len(values) == 0:
        raise ValueError("Categories and values cannot be empty")
    if len(categories) != len(values):
        raise ValueError("Number of categories must match number of values")

    # Create the bar chart
    plt.figure(figsize=(6, 4))
    plt.bar(categories, values, color=['red', 'blue', 'green', 'purple'])

    # Customize the plot
    plt.title("Bar Chart")
    plt.xlabel("Categories")
    plt.ylabel("Values")
    plt.grid(axis='y', linestyle='--', alpha=0.6)

    # Show the plot
    plt.show()

except Exception as e:
    print(f"Error: {e}")
```
