### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the line plot
try:
    x = np.linspace(0, 10, 11)
    if len(x) == 0:
        raise ValueError("X values cannot be empty")
    
    y = x ** 2
    if len(y) != len(x):
        raise ValueError("Y values must have the same length as X values")
    
    # Create the line plot
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, marker='o', linestyle='-', color='blue')

    # Customize the plot
    plt.title("Line Plot")
    plt.xlabel("X values")
    plt.ylabel("Y values")
    plt.grid(True, linestyle='--', alpha=0.6)

    # Show the plot
    plt.show()

except Exception as e:
    print(f"Error: {e}")
```
